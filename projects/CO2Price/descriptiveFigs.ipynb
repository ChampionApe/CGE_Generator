### Emissions and targets

Load model data:

```python
name = 'm'
dbs = {t: GpyDB(pickle_path = os.path.join(d['processedData'], f'{name}_{t}')) for t in range(1990, 2020)}
glbls = {t: loadFromPickle(f'glob_{t}', repo = d['processedData']) for t in range(1990, 2020)}
```

Extract data on historical emissions:

```python
M = glbls[1990].db['MData'].vals
Mnorm = (100 * M / glbls[1990].db['M1990'].vals).rename('Emissions')
```

Define dataframe with emissions targets:

```python
ts = [1990, 2005, 2011]
df   = pd.DataFrame({t: dbs[t].get('MTarget_NO_V1').loc[:2075] for t in ts}) * 100
dfv2 = pd.DataFrame({t: dbs[t].get('MTarget_NO_V2').loc[:2075] for t in ts}) * 100
```

Define the targets from 2020 and onward:

```python
target2020 = pd.concat([pd.Series(np.linspace(0.6, 0.3, 11), index = pd.Index(range(2020, 2031), name = 't')).iloc[0:-1],
                        pd.Series(np.linspace(0.3, 0, 16), index = pd.Index(range(2030, 2046), name = 't')).iloc[0:-1],
                        pd.Series(np.linspace(0, -0.1, 6), index = pd.Index(range(2045, 2051), name = 't'))],
                       axis = 0) * 100
target2020 = auxiliary.extrapolateUpper(target2020, 2075)
target2020_v2 = pd.concat([pd.Series(np.linspace(Mnorm.xs(2020)/100, 0.3, 11), index = pd.Index(range(2020, 2031), name = 't')).iloc[0:-1],
                        pd.Series(np.linspace(0.3, 0, 16), index = pd.Index(range(2030, 2046), name = 't')).iloc[0:-1],
                        pd.Series(np.linspace(0, -0.1, 6), index = pd.Index(range(2045, 2051), name = 't'))],
                       axis = 0) * 100
target2020_v2 = auxiliary.extrapolateUpper(target2020_v2, 2075)
```

Add to the dataframe:

```python
df[2020] = target2020
dfv2[2020] = target2020_v2
```

Plot:

```python
mult_graphs()
nplots = 2
nrows = math.ceil(nplots/2)
fig, axes = plt.subplots(nrows, min(nplots, 2), figsize = (14, (5*nrows)));

ax = plt.subplot(nrows, min(nplots,2), 1)
seaborn.lineplot(data = df, ax = ax, linewidth = 3, legend=False)
Mnorm.plot(ax=ax,linewidth=1.5, color='k',  legend=False)
ax.set_xlabel('Years', labelpad=10);
ax.set_xlim([1990, 2075]);
ax.set_ylabel('Emissions, index 100 = 1990', labelpad =10);
ax.set_ylim([-10, 130]);
ax.axhline(y=0, linewidth=1, color = 'k', alpha = 0.5);
ax.set_title('Version 1')

ax = plt.subplot(nrows, min(nplots,2), 2)
seaborn.lineplot(data = dfv2, ax = ax, linewidth = 3, legend=False)
Mnorm.plot(ax=ax,linewidth=1.5, color='k', legend=False)
ax.set_xlabel('Years', labelpad=10);
ax.set_xlim([1990, 2075]);
ax.set_ylabel('Emissions, index 100 = 1990', labelpad =10);
ax.set_ylim([-10, 130]);
ax.axhline(y=0, linewidth=1, color = 'k', alpha = 0.5);
ax.set_title('Version 2')

fig.legend(df.columns.insert(len(df.columns),Mnorm.name), loc=9,ncol=5,frameon=True)
fig.tight_layout();
fig.subplots_adjust(top=0.8);
fig.savefig(f"{d['figs']}\\TargetPaths_NoOvershoot.pdf",edgecolor='k')
```

### Version 1: Effective single year targets

We distinguish between (1) frozen policy regime and (2) constant reduction path policy.

a) Frozen policy version:

```python
T = 2075
# Regime 1:
rr = 20/15 # reduction rate, percentage point reduction per year
s1_FP = pd.concat([pd.Series(None, index = pd.Index(range(1990,2005), name = 't'), dtype = 'object'),
                    pd.Series(0.8, index = pd.Index(range(2005, T+1), name = 't'))], axis = 0).rename('Frozen policy') * 100
s1_CP = pd.concat([pd.Series(None, index = pd.Index(range(1990,2005), name = 't'), dtype = 'object'),
                   pd.Series(np.linspace(0.8, 0, 61), index = pd.Index(range(2005, 2066), name = 't')),
                   pd.Series(0, index = pd.Index(range(2066,T+1), name = 't'))], axis = 0).rename('Fixed reduction rate') * 100
# Regime 2:
rr = 1/6
s2_FP = pd.concat([pd.Series(None, index = pd.Index(range(2005,2011), name = 't'), dtype = 'object'),
                   pd.Series(0.79, index = pd.Index(range(2011, T+1), name = 't'))], axis = 0).rename('Frozen policy') * 100
s2_CP = pd.concat([pd.Series(None, index = pd.Index(range(2005,2011), name = 't'), dtype = 'object'),
                   pd.Series(np.linspace(0.79, 0.68333, T+1-2011), index = pd.Index(range(2011, T+1), name = 't'))], axis = 0).rename('Fixed reduction rate') * 100
rr = (79-60)/(2020-2011)
s3_FP = pd.concat([pd.Series(None, index = pd.Index(range(2011, 2020), name = 't'), dtype = 'object'),
                   pd.Series(0.60, index = pd.Index(range(2020, T+1), name = 't'))], axis = 0).rename('Frozen policy') * 100
s3_CP = pd.concat([pd.Series(None, index = pd.Index(range(2011, 2020), name = 't'), dtype = 'object'),
                   pd.Series(np.linspace(0.60, 0.00899, 29), index = pd.Index(range(2020, 2049), name = 't')),
                   pd.Series(0, index = pd.Index(range(2049, T+1), name = 't'))], axis = 0).rename('Fixed reduction rate') * 100

s4_FP = pd.concat([pd.Series(None, index = pd.Index(range(2020, 2030), name = 't'), dtype='object'),
                   pd.Series([0.3, 0], index = pd.Index([2030, 2045], name = 't'), dtype = 'object'),
                   pd.Series(-.1, index = pd.Index(range(2050, T+1), name = 't'), dtype='object')], axis = 0).rename('Frozen policy') * 100
```

Print target:

```python
df1 = pd.concat([s1_FP, s1_CP, Mnorm.loc[1990:2005].rename('Emissions')], axis=1)
df2 = pd.concat([s2_FP, s2_CP, Mnorm.loc[1990:2011].rename('Emissions')], axis=1)
df3 = pd.concat([s3_FP, s3_CP, Mnorm.loc[1990:2020].rename('Emissions')], axis=1)
df4 = pd.concat([s4_FP, Mnorm.loc[1990:2021].rename('Emissions')], axis = 1)
```

The four regimes:

```python
mult_graphs();
fig, axes = plt.subplots(2,2, figsize  = (14,8));
ax = plt.subplot(2,2,1);
seaborn.scatterplot(data = df1, ax = ax);
ax.axhline(y=0, linewidth = 1, color = 'k', alpha = 0.5);
ax.set_xlim([1990, T]);
ax.set_ylim([-10, 130]);
ax.set_xlabel(None, labelpad = 10);
ax.set_ylabel('Emission target', labelpad=10);
ax.set_title('For $t< 2005$');

ax = plt.subplot(2,2,2);
seaborn.scatterplot(data = df2, ax = ax);
ax.axhline(y=0, linewidth = 1, color = 'k', alpha = 0.5);
ax.set_xlim([1990, T]);
ax.set_ylim([-10, 130]);
ax.set_xlabel(None, labelpad = 10);
ax.set_ylabel('Emission target', labelpad=10);
ax.set_title('For $2005\leq t < 2011)$');

ax = plt.subplot(2,2,3);
seaborn.scatterplot(data = df3, ax = ax);
ax.axhline(y=0, linewidth = 1, color = 'k', alpha = 0.5);
ax.set_xlim([1990, T]);
ax.set_ylim([-10, 130]);
ax.set_xlabel(None, labelpad = 10);
ax.set_ylabel('Emission target', labelpad=10);
ax.set_title('For $2011\leq t < 2020$');

ax = plt.subplot(2,2,4);
seaborn.scatterplot(data = df4, ax = ax);
ax.axhline(y=0, linewidth = 1, color = 'k', alpha = 0.5);
ax.set_xlim([1990, T]);
ax.set_ylim([-10, 130]);
ax.set_xlabel(None, labelpad = 10);
ax.set_ylabel('Emission target', labelpad=10);
ax.set_title('For $t\geq 2020$');
fig.tight_layout()
# fig.legend(df1.columns,loc=8,ncol=3,frameon=True)
# fig.subplots_adjust(bottom=0.12);
fig.savefig(f"{d['figs']}\\SingleYearFP.pdf",edgecolor='k')
```
