```python
import numpy as np, scipy, pandas as pd
```

*Plot effective CO2 tax with/without EOP abatement (single technology, normally distributed costs):*

```python
c = 5 # average costs
σ = 2 # std. deviation of costs
τ = np.linspace(-10, 10, 41) # grid of taxes to compute costs on
```

Distribution

```python
Φ = scipy.stats.norm.cdf((τ-c)/σ)
ϕ = scipy.stats.norm.pdf((τ-c)/σ)
AvgTechCost = (c-σ*ϕ/Φ)
tc = τ*(1-Φ)+Φ*AvgTechCost
(tc-τ).max() # should always be negative (or zero)
```

Plot costs with/without:

```python
pd.DataFrame({'Tax': τ, 'TC': tc}).plot()
```

Now with multiplicative costs:

```python
σ_mult = σ/c
Φ_mult = scipy.stats.norm.cdf((τ/c-1)/σ_mult)
ϕ_mult = scipy.stats.norm.pdf((τ/c-1)/σ_mult)
AvgTechCost = c * (1 - σ_mult * ϕ_mult/Φ_mult)
tc_mult = τ *(1-Φ_mult)+Φ_mult*AvgTechCost
(tc_mult-τ).max()
```
